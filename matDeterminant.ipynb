```python
#rank and determinant
import numpy as np
mat=np.random.randn(2,2)
m=300
mat2=np.random.randn(m,m)

mat2[:,2]=6*mat2[:,0]
print(np.linalg.matrix_rank(mat2))
print(np.linalg.det(mat2))
```

```python
#effect of row swapping on determinant
matrix1=np.random.randn(6,6)
print(np.linalg.det(matrix1))
temp=np.array(matrix1[1,:])               #swapping rows
matrix1[1,:]=matrix1[0,:]
matrix1[0,:]=temp
print(np.linalg.det(matrix1))
```

```python
#effect of row swapping on determinant
matrix1=np.random.randn(6,6)

print(np.linalg.det(matrix1))
matrix1[[0,2],:]=matrix1[[2,0],:]              #swapping rows
print(np.linalg.det(matrix1))
matrix1[[4,5],:]=matrix1[[5,4],:]              #swapping more rows
print(np.linalg.det(matrix1))
```

```python
#determinant of shifted matrix
import matplotlib.pyplot as plt
lambda1=np.linspace(0,0.1,30)


tmp=np.zeros(1000)
dets=np.zeros(len(lambda1))


for deti in range(len(lambda1)):

    #run for 1000 iterations

    for i in range(1000):
        M=np.random.randn(20,20)
        
        M[:,0]=M[:,1] # imposing linear dependency
        
        tmp[i]=abs(np.linalg.det(M+lambda1[deti]*np.eye(20)))  # shifting a matrix
    
    dets[deti]=np.mean(tmp)

plt.plot(lambda1,dets,'s-')
```

```python
#illustrate det(a)*det(b)=det(ab)
A=np.random.randn(3,3)
B=np.random.randn(3,3)
AB=A@B
dets=np.zeros((40,2))
print(np.linalg.det(A)*np.linalg.det(B),np.linalg.det(AB))
for i in range(40):
    A=np.random.randn(i,i)
    B=np.random.randn(i,i)
    AB=A@B
    dets[i,:]=[np.linalg.det(A)*np.linalg.det(B),np.linalg.det(AB)]

plt.plot(dets[:,0]-dets[:,1],'s-') #plotting the differnece 
plt.ylim([1,-1])
```
