```python
import numpy as np
import matplotlib.pyplot as plt
v1=np.array([1,2,3,4,5,6])

#vector length ‖v‖=✓v.v
vl1=np.sqrt(np.sum(np.multiply(v1,v1)))

vl2=np.linalg.norm(v1)
print(vl1,vl2)
```

```python
v1=np.array([2,4,-3])
v2=np.array([0,-3,-3])

#calculating angle between vectors
#using law of cosine proved aTb=‖a‖‖b‖cos(θ)

ang=np.arccos(np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)))


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'b')
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'r')

plt.axis((-6, 6, -6, 6))
plt.title('Angle between vectors: %s rad.' %ang)
plt.show()
```

```python
v1=np.array([2,4,-3])
v2=np.array([0,-3,-3])

#algebraic formula 
dp_a=np.dot(v1,v2)

#geometric
dp_b=np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(ang)
print(dp_a,dp_b)
```

```python
v1=np.array([1,2,3])
v2=np.array([-1,0,1])

#outer product
print(np.outer(v1,v2))

op=np.zeros((len(v1),len(v2)))
for i1,i in enumerate(v1):
    for j1,j in enumerate(v2):
        op[i1,j1]=i*j

print(op)
```

```python
z=complex(3,4)

#magnitude
print(np.linalg.norm(z))

#using regular transpose 
print(np.transpose(z)*z)

#using hermition transpose
print(np.transpose(z.conjugate())*z)

# complex vector
v = np.array( [ 3, 4j, 5+2j, complex(2,-5) ] )
print( v.T)
print( np.transpose(v) )
print( np.transpose(v.conjugate()) )
```

```python
v1=np.array([-.3,.6])
mu=1/np.linalg.norm(v1)

v1n = v1*mu

# plot them
plt.plot([0, v1n[0]],[0, v1n[1]],'r',label='v1-norm',linewidth=5)
plt.plot([0, v1[0]],[0, v1[1]],'b',label='v1')

# axis square
plt.axis('square')
plt.axis(( -6, 6, -6, 6 ))
plt.grid()
plt.legend()
plt.show()
```
